```python
beta = 0.85
N = urls.shape[0]
#m = pd.DataFrame(data=urls['ID'])
m = pd.DataFrame(0, index=urls['ID'].tolist(), columns=urls['ID'])

outLinks = pd.DataFrame(0, columns=['count'], index=urls['ID'].tolist())

def calculateOutLinks(link):
    origin = link['fromlink']
    outLinks['count'][origin] += 1

links.apply(calculateOutLinks, axis=1)

outLinks
```

```python
def calculateM(link):
    i = link['fromlink']
    j = link['tolink']
    m.ix[j,i] = 1.0/outLinks['count'][i]

links.apply(calculateM, axis=1)

m
```

```python
r = pd.DataFrame(1.0/N, columns=['count'], index=urls['ID'].tolist())
r
```

```python
a = beta * m + (1.0-beta)/N
a
```

```python
converged = False 

oldR = pd.DataFrame(0, columns=['count'], index=urls['ID'].tolist())

while(not converged): 
    oldR = r 
    r = a.dot(r) 
    difference = r-oldR 
    if r.equals(oldR): 
        converged = True 

r
```

```python
r.sort_values(by = 'count', ascending = False, inplace=True)
top10 = r[:10]

indexedUrls = urls.set_index('ID')

def printURL(pagerank):
    print (indexedUrls['URL'][pagerank.name])
    
top10.apply(printURL, axis=1)

0
```
